This assignment demonstrates the creation and visualization of a simple graph using the NetworkX package in Python. The graph used is the Krackhardt Kite Graph, a well-known social network example consisting of 10 individuals connected by social relationships.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Creating the Graph

The graph consists of ten nodes representing individuals and edges representing relationships between them.

```python
G = nx.Graph()

nodes = [
    "Andre", "Beverly", "Carol", "Diane", "Ed",
    "Fernando", "Garth", "Heather", "Ike", "Jane"
]

G.add_nodes_from(nodes)

edges = [
    ("Andre", "Beverly"),
    ("Andre", "Carol"),
    ("Andre", "Diane"),
    ("Andre", "Fernando"),
    ("Beverly", "Diane"),
    ("Beverly", "Ed"),
    ("Beverly", "Garth"),
    ("Carol", "Diane"),
    ("Carol", "Fernando"),
    ("Diane", "Ed"),
    ("Diane", "Fernando"),
    ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"),
    ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

G.add_edges_from(edges)
```

## Visualize the Graph

```python
plt.figure(figsize=(10,7))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="skyblue",
    node_size=2500,
    font_size=10,
    font_weight="bold"
)

plt.title("Krackhardt Kite Graph")
plt.show()
```

## Basic Graph Statistics

```python
print("Number of Nodes:", G.number_of_nodes())
print("Number of Edges:", G.number_of_edges())
```

```python
print("\nDegree of Each Node")

for node, degree in G.degree():
    print(node, ":", degree)
```
