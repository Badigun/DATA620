# Project Overview

Professional basketball is built on relationships between players, teams, and seasons. As players move between teams throughout their careers, they create connections with former teammates, forming a large network that spans the entire NBA. Network analysis provides an effective way to study these relationships and identify the players who have the greatest influence within the league.

The objective of this project is to construct and analyze an NBA player network based on shared team memberships. Two players will be connected if they played for the same team during the same season. By examining this network, I hope to identify the most connected players, discover communities of teammates, and better understand how player movement shapes the structure of the NBA.

### Guiding Question

> Which NBA players occupy the most influential positions in the league based on their connections to teammates through shared team memberships?

### Hypothesis

Players who have long careers or have played for multiple franchises will have higher centrality measures because they are connected to more teammates and serve as bridges between different player communities.

# Data Source

The project will use publicly available NBA player and roster data from **Kaggle** or **Basketball-Reference**.

The dataset will include information such as:

- Player names
- NBA teams
- Seasons played
- Team rosters

The network will be constructed as follows:

- Each **node** represents an NBA player.
- An **edge** connects two players who played on the same team during the same season.
- Edge weights may represent the number of seasons two players were teammates.

# Plan for the Analysis

The project will be completed using **Python** in a **Jupyter Notebook**.

The planned workflow includes:

1. Import and clean the NBA roster dataset.
2. Create a player-to-player network based on shared team memberships.
3. Calculate descriptive network statistics.
4. Compute centrality measures, including:
   - Degree Centrality
   - Betweenness Centrality
   - Closeness Centrality
   - Eigenvector Centrality
5. Detect player communities using community detection algorithms.
6. Visualize the NBA player network.
7. Identify the most influential players based on multiple centrality measures.
8. Interpret the findings and discuss how player movement influences the structure of the NBA network.

# Expected Outcomes

The analysis is expected to identify:

- Players with the highest degree centrality who have played with many teammates.
- Players with high betweenness centrality who connect different player communities after changing teams.
- Communities of players formed through long-term team relationships.
- Network structures that illustrate how player movement links franchises across the NBA.

These findings will demonstrate how network analysis can reveal patterns that are not immediately visible through traditional basketball statistics.
